# load Fifa data

```python
#read data from sheet fifa_data
fifa=pd.read_csv("fifa_data.csv")
fifa.head()
```

# Histograms

```python
#Histograms chart for overall
bins=[40,50,60,70,80,90,100]
#Histograms chart for overall
#plt.hist(V.columns,color="g" )   
plt.hist(fifa.Overall,bins=bins,color="g" )     
plt.title("Distribution of player skills on FIFA 2018")
plt.xlabel("Skill level")

plt.ylabel("Number of player")
plt.show()
```

# Pie chart

```python
#تحويل الاسماء الى ارقام 
Left=fifa.loc[fifa['Preferred Foot']=='Left'].count()[0]
right=fifa.loc[fifa['Preferred Foot']=='Right'].count()[0]
#labels اسماء القيم 
#color لوان القيم 
#تغير شكل الرسم
#plt.style_use("default")
labels=['Left','Right']
colors=['red','b']

plt.title("foot Preferred of fifa player ")
#autopct='%.2f %%') اضافة النسبة الى الرسم
#تفكيك الجدول 
#explode=(1,1)
#explode=explode

#labels['V1','V2']
#colors['V1','V2']
#plt.pie([V2,V2],labels=labels,colors=colors ,autopct='%.2f %%'النسبة المكتوبة )

plt.pie([Left,right],labels=labels,colors=colors ,autopct='%.2f %%' )
plt.show()
```
